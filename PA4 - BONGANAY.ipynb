#### 1. Create the following data frames based on the format provided:

a. Filename: Instru = [“Name”, “GEAS”, “Electronics >70”]; where track is constant as
Instrumentation and hometown Luzon

```python
import pandas as pd

df = pd.read_csv("board2.csv")

Instru = df[(df["Track"] == "Instrumentation") & 
            (df["Hometown"] == "Luzon") & 
            (df["Electronics"] > 70)] [["Name", "GEAS", "Electronics"]]

print("INSTRU DATAFRAME:")
print(" ")
Instru
             
```

b. Filename: Mindy = [ “Name”, “Track”, “Electronics”, “Average >=55”]; where hometown is
constant as Mindanao and gender Female

```python
import pandas as pd

df = pd.read_csv("board2.csv")
df["Average"] = df[["Math", "Electronics", "GEAS", "Communication"]].mean(axis=1)

Mindy = df[(df["Hometown"] == "Mindanao") &
            (df["Gender"] == "Female") & 
            (df["Average"] >= 55)] [["Name", "Track", "Electronics", "Average"]]

print("MINDY DATAFRAME:")
print(" ")
Mindy
         
```

#### 2. Create a visualization that shows how the different features contributes to average grade. 
Does chosen track in college, gender, or hometown contributes to a higher average score?


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("board2.csv")
df["Average"] = df[["Math", "Electronics", "GEAS", "Communication"]].mean(axis=1)

plt.figure(figsize=(6,6))
sns.barplot(data=df, x="Gender", y="Average", estimator="mean", errorbar=None)
plt.title("Average Score by Gender")
plt.ylabel("Mean Average Score")
plt.xlabel("Gender")
plt.show()

plt.figure(figsize=(8,6))
sns.barplot(data=df, x="Track", y="Average", estimator="mean", errorbar=None)
plt.title("Average Score by College Track")
plt.ylabel("Mean Average Score")
plt.xlabel("Track")
plt.show()

plt.figure(figsize=(8,6))
sns.barplot(data=df, x="Hometown", y="Average", estimator="mean", errorbar=None)
plt.title("Average Score by Hometown")
plt.ylabel("Mean Average Score")
plt.xlabel("Hometown")
plt.show()
```
